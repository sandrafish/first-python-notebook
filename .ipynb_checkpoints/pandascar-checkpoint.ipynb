```python
top_supporters = support.groupby(["contributor_firstname", "contributor_lastname"]
).amount.sum().reset_index().sort_values("amount", ascending=False).head(10)
```

```python
top_supporters.head(10)
```

```python
top_supporters.amount.plot.bar()
```

```python
top_supporters.amount.plot.barh()
```

```python
top_supporters.head(5).amount.plot.barh()
```

```python
#creating a new column that concats first and last
top_supporters['contributor_fullname'] = top_supporters.contributor_firstname + " " + top_supporters.contributor_lastname
```

```python
top_supporters.head(5)
```

```python
chart = top_supporters.head(5).amount.plot.barh()
```

```python
chart = top_supporters.head(10).amount.plot.barh()
chart.set_yticklabels(top_supporters.contributor_fullname)
```

```python
top_supporters.head(10).to_csv("top_supporters.csv")
```

```python
top_opponents = oppose.groupby(["contributor_firstname", "contributor_lastname"]
).amount.sum().reset_index().sort_values("amount", ascending=False).head(10)
```

```python
top_opponents.head(5)
```

```python
top_opponents.head(10).to_csv("top_opponents.csv")
```

```python
#creating a new column that concats first and last
top_opponents['contributor_fullname'] = top_opponents.contributor_firstname + " " + top_opponents.contributor_lastname
```
